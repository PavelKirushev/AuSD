1) При помощи стека (можно использовать любую реализацию стека, например, из лекции) проверить, что в строке, содержащей большой фрагмент кода на python, корректно открываются и закрываются скобки '(' и ')'. Заводить какие-либо счетчики и вообще вести подсчет скобок запрещено. 
Например, в следующей строке корректное количество скобок: 
'((dsasdas((das(d)das)d)asd)da)' 
А вот в следующей закрывающих больше, чем открывающих:
'(f(gfhd((hgfdg((fghh)dads)das))ds)adddads())))' 

```
s = input()
s1 = ""
for i in range(len(s)):
    if s[i] == '(' or s[i] == ')':
        s1 += s[i]
s = s1
stack = []
f = 1
for i in range(len(s)):
    if s[i] == "{" or s[i] == "(" or s[i] == "[":
        stack.append(s[i])
    else:
        if not stack:
            f = 0
            break
        l = stack[len(stack) - 1]
        if l == "(" and s[i] == ")" or l == "[" and s[i] == "]" or l == "{" and s[i] == "}":
            stack.pop()
        else:
            break

if not stack and f:
    print("True")
else:
    print("False")
```

2) С помощью списка (list) реализовать стек. При создании стека нужно передать максимальную длину стека. В случае попытки добавления элемента, приводящего к превышению емкости стека возвращать ошибку. 
Стек должен поддерживать следующие операции: 
S.push(e): добавление элемента e на вершину стека S. 
S.pop(): удаляет и возвращает верхний элемент стека S. Если стек пуст, то возникает ошибка. 
S.top(): возвращает верхний элемент стека S, не удаляя его. Если стек пуст, то возникает ошибка. 
S.is_empty( ): возвращает True если стек S не содержит ни одного элемента. 
len(S): возвращает текущеее количество элементов в стеке S. 

```
class Stack:
    def __init__(self, n) -> None:
        self.items = [None] * n
        self.ukaz = 0

    def push(self, e):
        if self.ukaz == len(self.items):
            print("error")
            return
        self.items[self.ukaz] = e
        self.ukaz = self.ukaz + 1

    def pop(self):
        if self.ukaz == 0:
            print("error")
            return
        self.items[self.ukaz - 1] = None
        self.ukaz = self.ukaz - 1

    def top(self):
        if self.ukaz == 0:
            print("error")
            return
        return self.items[self.ukaz - 1]

    def is_empty(self):
        if self.ukaz == 0:
            return True
        return False
ll = Stack(5)
ll.push(8)
ll.push(5)
ll.push(11)
ll.push(9)
ll.push(91)
ll.pop()
ll.push(5)
ll.pop()
ll.pop()
ll.push(3)
print(ll.top())
print(ll.items)
```

3) Реализовать класс однонаправленного связанного списка.
3.1) Реализовать метод добавления элемента в начало списка.  

```
class Node:
    def __init__(self, val: int, next = None):
        self.val = val
        self.next = next

class LinkedList:
    def __init__(self, head) -> None:
        self.head = head

    def insert(self, x):
        tmp = self.head
        while tmp.next:
            tmp = tmp.next
        tmp.next = Node(val=x)

    def insert_to_start(self, x):
        tmp = self.head
        current = Node(val=x)
        self.head = current
        current.next = tmp

    def delete(self, index):
        tmp = self.head
        prev = None
        num = 1
        while num != index:
            num += 1
            prev = tmp
            tmp = tmp.next
        prev.next = tmp.next

    def output(self):
        tmp = self.head
        while tmp:
            print(tmp.val, end=" ")
            tmp = tmp.next
        print("")

head = Node(10)
ll = LinkedList(head=head)
ll.insert(15)
ll.insert(30)
ll.insert(45)
ll.insert(60)
ll.insert_to_start(-5)
ll.output()
ll.delete(3)
ll.output()

```

4) На базе класса однонаправленного связанного списка реализовать двунаправленный связанный список.  
4.1) Реализовать метод добавления элемента в начало списка.  

```
class Node:
    def __init__(self, val: int, next = None, prev = None):
        self.val = val
        self.next = next
        self.prev = prev

class LinkedList:
    def __init__(self, head) -> None:
        self.head = head
    def insert(self, x):
        tmp = self.head
        while tmp.next:
            tmp = tmp.next
        tmp.next = Node(val=x, prev=tmp)

    def insert_to_start(self, x):
        tmp = self.head
        self.head = Node(val=x, next=tmp)

    def delete(self, index):
        tmp = self.head
        num = 1
        while num != index:
            num += 1
            tmp = tmp.next
        tmp.prev.next = tmp.next

    def output(self):
        tmp = self.head
        while tmp:
            print(tmp.val, end=" ")
            tmp = tmp.next
        print("")

head = Node(10)
ll = LinkedList(head=head)
ll.insert(15)
ll.insert(30)
ll.insert(45)
ll.insert(60)
ll.insert_to_start(-5)
ll.output()
ll.delete(3)
ll.output()
```

5) При помощи класса связанного списка реализовать очередь.

```
class Node:
    def __init__(self, val: int, next = None, prev = None):
        self.val = val
        self.next = next
        self.prev = prev

class LinkedList:
    def __init__(self, head) -> None:
        self.head = head
    def insert(self, x):
        tmp = self.head
        while tmp.next:
            tmp = tmp.next
        tmp.next = Node(val=x, prev=tmp)

    def pop(self):
        self.head = self.head.next

    def output(self):
        tmp = self.head
        while tmp:
            print(tmp.val, end=" ")
            tmp = tmp.next
        print("")

head = Node(10)
ll = LinkedList(head=head)
ll.insert(15)
ll.insert(30)
ll.insert(45)
ll.insert(60)
ll.output()
ll.pop()
ll.pop()
ll.output()
```
