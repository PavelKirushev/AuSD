Необходимо отсортировать список товаров по количеству продаж и вывести результат на экран. В зависимости от переданного параметра отсортировать список товаров по возрастанию или по убыванию количества продаж, используя алгоритмы сортировки: сортировку вставками, сортировку выбором и быструю сортировку. Сравнить время выполнения алгоритмов сортировки с помощью декоратора. Данные о товарах хранятся в файле.

```python
import time
import random

def timer_decorator(func):
    def wrapper(*args, **kwargs):
        if not wrapper.has_been_called:
            wrapper.has_been_called = True
            start_time = time.time()
            result = func(*args, **kwargs)
            end_time = time.time()
            print(f"Время выполнения {func.__name__}: {end_time - start_time} секунд")
            wrapper.has_been_called = False
            return result
        else:
            return func(*args, **kwargs)
    wrapper.has_been_called = False
    return wrapper

@timer_decorator
def vstavka(mas):
    for i in range(len(mas) - 1):
        item = mas[i]
        j = i
        while j > 0 and item < mas[j - 1]:
            mas[j] = mas[j - 1]
            j -= 1
        mas[j] = item
    return mas

@timer_decorator
def vibor(mas):
    for i in range(len(mas) - 1):
        min_in = i
        for j in range(i + 1, len(mas)):
            if mas[j] < mas[min_in]:
                min_in = j
        mas[i], mas[min_in] = mas[min_in], mas[i]
    return mas

@timer_decorator
def quick(mas):
    if len(mas) < 2:
        return mas
    else:
        pivot = mas[0]
        less = [i for i in mas[1:] if i <= pivot]
        greater = [i for i in mas[1:] if i > pivot]
        return quick(less) + [pivot] + quick(greater)

mas = [random.randint(1, 10000) for _ in range(1000)]
vstavka(mas)
vibor(mas)
quick(mas)
```
