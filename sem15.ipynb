11. Класс – целочисленный вектор фиксированной размерности. Методы – сумма, разность, скалярное произведение векторов. Поля – размерность вектора, координаты вектора.

```python
class Vector:
    def __init__(self, x: int, y: int):
        self.x = x
        self.y = y
        self.len = (x ** 2 + y ** 2) ** 0.5

    def summa(self, vector2):
        if not isinstance(vector2, Vector):
            print("error")
            return
        new_x = self.x + vector2.x
        new_y = self.y + vector2.y
        new_vector = Vector(new_x, new_y)
        return new_vector

    def raznost(self, vector2):
        if not isinstance(vector2, Vector):
            print("error")
            return None
        new_x = self.x - vector2.x
        new_y = self.y - vector2.y
        new_vector = Vector(new_x, new_y)
        return new_vector

    def scalyar(self, vector2):
        if not isinstance(vector2, Vector):
            print("error")
            return None
        return (self.x * vector2.x + self.y * vector2.y)

vector1 = Vector(2, 1)
vector2 = Vector(2, 2)
vector3 = vector1.summa(vector2)
print(vector1.scalyar(vector2))
if vector3:
    print(vector3.len)
```

11. Опишите класс Date, заданный тремя атрибутами для года, месяца и дня. Включите в описание класса методы, позволяющие вычислить дату следующего дня, определить, сколько дней осталось до конца месяца, и свойство, позволяющее выяснить, является ли год високосным.

```python
class Date:
    def __init__(self, day, month, year):
        self.d = day
        self.m = month
        self.y = year

    def nextDay(self):
        if self.d < 28:
            return Date(self.d + 1, self.m, self.y)
        if self.m == 2 and self.leapYear and self.d == 28:
            return Date(self.d + 1, self.m, self.y)
        if (self.m % 2 == 1 and self.m < 8 or self.m % 2 == 0 and self.m > 7) and self.d < 31:
            return Date(self.d + 1, self.m, self.y)
        if (self.m % 2 == 0 and self.m < 8 and self.m != 2 or self.m % 2 == 1 and self.m > 7) and self.d < 30:
            return Date(self.d + 1, self.m, self.y)
        else:
            if self.m < 12:
                return Date(1, self.m + 1, self.y)
            else:
                return Date(1, 1, self.y + 1)

    @property
    def leapYear(self):
        if self.y % 400 == 0 or self.y % 4 == 0 and self.y % 100 != 0:
            return True
        return False

    def daysLeft(self):
        if self.m % 2 == 1 and self.m < 8 or self.m % 2 == 0 and self.m > 7:
            print(f"{31 - self.d} days until end of month")
        elif self.m == 2 and self.leapYear:
            print(f"{29 - self.d} days until end of month")
        elif self.m == 2 and not self.leapYear:
            print(f"{28 - self.d} days until end of month")
        else:
            print(f"{30 - self.d} days until end of month")
        return

    def output(self):
        print(f"{self.d}.{self.m}.{self.y}")

current_day = Date(31, 12, 2000)
current_day.nextDay().output()
print(current_day.leapYear)
current_day.daysLeft()
```

Создайте класс ТЕЛО с методами вычисления площади поверхности и объема, а также методом, выводящим информацию о фигуре на экран. Создайте дочерние классы ПАРАЛЛЕЛЕПИПЕД, ШАР, ПИРАМИДА со своими методами вычисления площади и объема. Создайте список  n  фигур и выведите полную информацию о фигурах на экран.

```python
class Body:
    def __init__(self):
        pass

    def surface_area(self):
        raise NotImplementedError("Subclass must implement abstract method")

    def volume(self):
        raise NotImplementedError("Subclass must implement abstract method")

    def info(self):
        raise NotImplementedError("Subclass must implement abstract method")


class Parallelepiped(Body):
    def __init__(self, length, width, height):
        self.len = length
        self.wid = width
        self.hei = height

    def surface_area(self):
        return 2 * (self.len * self.wid + self.wid * self.hei + self.len * self.hei)

    def volume(self):
        return self.len * self.wid * self.hei

    def info(self):
        return f"Parallelepiped: Length = {self.len}, Width = {self.wid}, Height = {self.hei}"

class Sphere(Body):
    def __init__(self, rad):
        self.rad = rad

    def surface_area(self):
        return round(4 * 3.14159 * self.rad ** 2, 4)

    def volume(self):
        return round(4 * 3.14159 * self.rad ** 2, 4)

    def info(self):
        return f"Sphere: Radius = {self.rad}"


class Pyramid(Body):
    def __init__(self, length, width, height):
        self.len = length
        self.wid = width
        self.hei = height

    def surface_area(self):
        osn = self.len * self.wid
        bok = 0.5 * self.hei * (self.len + self.wid)
        return osn + bok

    def volume(self):
        return 1/3 * self.len * self.wid * self.hei

    def info(self):
        return f"Pyramid: Length = {self.len}, Width = {self.wid}, Height = {self.hei}"

figures = [
    Parallelepiped(3, 4, 5),
    Sphere(2),
    Pyramid(4, 3, 6)
]
for figure in figures:
    print(figure.info())
    print(f"Surface Area: {figure.surface_area()}")
    print(f"Volume: {figure.volume()}")
    print()
```
