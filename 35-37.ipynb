Создайте двусвязный список для хранения информации о студентах в университете. Каждый элемент списка должен содержать имя, фамилию, номер группы, список предметов, которые студент изучает, и оценки по каждому предмету.

```python
class Node():
    def __init__(self, firstname: str, secondname: str, numgroup: str, list: list, marks: list, next=None, prev=None):
        self.next = next
        self.prev = prev
        self.first = firstname
        self.second = secondname
        self.num = numgroup
        self.list = list
        self.marks = marks

class LinkedList():
    def __init__(self, firstname, secondname, numgroup, list, marks):
        self.head = Node(firstname, secondname, numgroup, list, marks)
        self.tail = self.head
        self.size = 1

    def pushback(self, firstname, secondname, numgroup, list, marks):
        if self.size == 0:
            return Node(firstname, secondname, numgroup, list, marks)
        tmp = self.tail
        tmp.next = Node(firstname, secondname, numgroup, list, marks, prev=tmp)
        self.tail = tmp.next
        self.size += 1

    def delete(self):
        if self.size <= 0:
            print("ERROR")
            return
        if self.size == 1:
            self.head = None
            self.tail = None
            self.size = 0
            return
        tmp = self.tail.prev
        self.tail = tmp
        self.tail.next = None
        self.size -= 1
        return

    def size(self):
        return self.size

    def output(self):
        tmp = self.head
        while tmp:
            print(f"Имя: {tmp.first}")
            print(f"Фамилия: {tmp.second}")
            print(f"Номер группы: {tmp.num}")
            print(f"Список предметов: ", end='')
            for i in tmp.list:
                print(f"{i}", end=' ')
            print(f"\nОценки: ", end='')
            for i, j in zip(tmp.list, tmp.marks):
                print(f"{i}-{j}", end=' ')
            print("\n")
            tmp = tmp.next
        print("\n")

ucheniki = LinkedList("Павел", "Кирушев", "ИД23-3", ["АИСД", "ТОД", "Практикум"], [2, 3, 2])
ucheniki.pushback("Денис", "Варламов", "ИД23-3", ["АИСД", "ТОД", "Практикум"], [2, 2, 2])
ucheniki.output()
ucheniki.delete()
ucheniki.pushback("Аркадий", "Фролкин", "ИД23-3", ["АИСД", "ТОД", "Практикум"], [1, 1, 1])
ucheniki.output()
```

Реализовать функцию, которая находит среднее арифметическое всех элементов двусвязного списка.

```python
class Node():
    def __init__(self, value, next=None, prev=None):
        self.next = next
        self.prev = prev
        self.value = value

class LinkedList():
    def __init__(self, value):
        self.head = Node(value)
        self.tail = self.head
        self.size = 1

    def pushback(self, value):
        if self.size == 0:
            return Node(value)
        tmp = self.tail
        tmp.next = Node(value, prev=tmp)
        self.tail = tmp.next
        self.size += 1

    def delete(self):
        if self.size <= 0:
            print("ERROR")
            return
        if self.size == 1:
            self.head = None
            self.tail = None
            self.size = 0
            return
        tmp = self.tail.prev
        self.tail = tmp
        self.tail.next = None
        self.size -= 1
        return

    def size(self):
        return self.size

    def output(self):
        tmp = self.head
        while tmp:
            print(tmp.value, end=' ')
            tmp = tmp.next
        print("")

    def output_reverse(self):
        tmp = self.tail
        while tmp:
            print(tmp.value, end=' ')
            tmp = tmp.prev
        print("")

    def srznach(self):
        tmp = self.head
        cnt = self.size
        if cnt == 0:
            return
        summa = 0
        while tmp:
            summa += tmp.value
            tmp = tmp.next
        return summa / cnt

spisok = LinkedList(5)
spisok.pushback(4)
spisok.pushback(6)
spisok.pushback(7)
spisok.pushback(3)
spisok.output()
spisok.output_reverse()
print(spisok.srznach())
```

Реализовать функцию, которая удаляет все повторяющиеся элементы из циклического двусвязного списка.

```python
class Node():
    def __init__(self, value, next=None, prev=None):
        self.next = next
        self.prev = prev
        self.value = value

class LinkedList():
    def __init__(self, value):
        self.head = Node(value)
        self.size = 1
        self.dic = {value: 1}

    def pushback(self, value):
        if self.size == 0:
            self.dic[value] = 1
            return Node(value)
        if self.size == 1:
            self.head.next = Node(value, prev=self.head, next=self.head)
            self.head.prev = self.head.next
            self.size += 1
            if value not in self.dic:
                self.dic[value] = 0
            self.dic[value] += 1
            return
        tmp = self.head.prev
        tmp.next = Node(value, prev=tmp, next=self.head)
        self.head.prev = tmp.next
        self.size += 1
        if value not in self.dic:
            self.dic[value] = 0
        self.dic[value] += 1

    def delete(self, value=None):
        if self.size <= 0:
            print("ERROR")
            return
        if self.size == 1:
            if not value:
                self.head = None
                self.size = 0
            else:
                if value != self.head:
                    print("error")
                else:
                    self.head = None
                    self.size = 0
            return
        if not value:
            self.dic[self.head.prev.value] -= 1
            self.head.prev.prev.next = self.head
        else:
            if value not in self.dic:
                print("error1")
                return
            elif self.dic[value] == 0:
                print("error2")
                return
            else:
                self.dic[value] -= 1
                tmp = self.head
                while tmp.value != value:
                    tmp = tmp.next
                tmp.prev.next = tmp.next
                tmp.next.prev = tmp.prev
        self.size -= 1
        return

    def size(self):
        return self.size

    def output(self):
        tmp = self.head
        while True:
            print(tmp.value, end=' ')
            tmp = tmp.next
            if tmp == self.head:
                break
        print("")

    def output_reverse(self):
        tmp = self.head.prev
        while True:
            print(tmp.value, end=' ')
            tmp = tmp.prev
            if tmp == self.head.prev:
                break
        print("")

    def povtor(self):
        for key, value in self.dic.items():
            while value > 1:
                self.delete(key)
                value -= 1


s = LinkedList(5)
s.pushback(4)
s.pushback(3)
s.pushback(6)
s.pushback(6)
s.pushback(4)
s.pushback(4)
s.output()
s.delete(3)
s.output()
s.povtor()
s.output()
```
